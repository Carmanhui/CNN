```python
from keras.preprocessing import image
import matplotlib.pyplot as plt #用于显式图片
```

```python
img = image.load_img('meinv.jpg',target_size=(150,150))
print(img.size)
plt.imshow(img)
plt.xlabel('width')    
plt.ylabel('height')    
plt.title('meinv') 
plt.text(100,20,'meinv',size=20,color='r')

plt.savefig('meinv1.jpg')
plt.show()
```

```python
target_size=(150,150)
img = img.resize(target_size)
print(img.size)
plt.imshow(img)
plt.show()
```
