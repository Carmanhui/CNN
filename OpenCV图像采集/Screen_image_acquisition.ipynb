```python
"""
对于屏幕上的某个区域进行截图，采集图像
"""
import os, sys
import cv2
import time
import PIL
from PIL import Image,ImageTk,ImageGrab
import win32gui, win32ui, win32con, win32api

if sys.version_info[0] == 2:
    from Tkinter import *
    from tkFont import Font
    from ttk import *
    #Usage:showinfo/warning/error,askquestion/okcancel/yesno/retrycancel
    from tkMessageBox import *
    #Usage:f=tkFileDialog.askopenfilename(initialdir='E:/Python')
    #import tkFileDialog
    #import tkSimpleDialog
else:  #Python 3.x
    from tkinter import *
    from tkinter.font import Font
    from tkinter.ttk import *
    from tkinter.messagebox import *
    import tkinter.filedialog as tkFileDialog
    #import tkinter.simpledialog as tkSimpleDialog    #askstring()

class Application_Screen_image_acquisition(Frame):
    #这个类仅实现界面生成功能，具体事件处理代码在子类Application中。
    def __init__(self, master=None):
        Frame.__init__(self, master)
        self.master.title('屏幕图像采集')
        a, b = self.main_window_info()
        self.master.geometry('560x376+%d+%d' % (a, b))
        self.createWidgets()
        
    def main_window_info(self):
        ws = self.winfo_screenwidth()
        hs = self.winfo_screenheight()
        x = (ws - 560) / 2
        y = (hs - 376) / 2
        return x, y

    def createWidgets(self):
        self.top = self.winfo_toplevel()

        self.style = Style()

        self.style.configure('TCommand4.TButton', font=('宋体',12))
        self.Command4 = Button(self.top, text='退出', command=self.Command4_Cmd, style='TCommand4.TButton')
        self.Command4.place(relx=0.771, rely=0.681, relwidth=0.173, relheight=0.088)

        self.style.configure('TCommand3.TButton', font=('宋体',12))
        self.Command3 = Button(self.top, text='采集图像', command=self.Command3_Cmd, style='TCommand3.TButton')
        self.Command3.place(relx=0.771, rely=0.489, relwidth=0.173, relheight=0.088)

        self.style.configure('TCommand1.TButton', font=('宋体',12))
        self.Command1 = Button(self.top, text='选择区域', command=self.Command1_Cmd, style='TCommand1.TButton')
        self.Command1.place(relx=0.771, rely=0.106, relwidth=0.173, relheight=0.088)

        self.style.configure('TLabel1.TLabel', anchor='w', foreground='#000000', background='#FFFFFF', font=('宋体',9))
        self.Label1 = Label(self.top, text='Label1', style='TLabel1.TLabel')
        self.Label1.place(relx=0.029, rely=0.043, relwidth=0.716, relheight=0.896)

        self.style.configure('TCommand2.TButton', font=('宋体',12))
        self.Command2 = Button(self.top, text='选择路径', command=self.Command2_Cmd, style='TCommand2.TButton')
        self.Command2.place(relx=0.771, rely=0.298, relwidth=0.173, relheight=0.088)

    def window_capture(self,filename,srcbmp=[0, 0, None, None]):
        hwnd = 0 # 窗口的编号，0号表示当前活跃窗口
        # 根据窗口句柄获取窗口的设备上下文DC（Divice Context）
        hwndDC = win32gui.GetWindowDC(hwnd)
        # 根据窗口的DC获取mfcDC
        mfcDC = win32ui.CreateDCFromHandle(hwndDC)
        # mfcDC创建可兼容的DC
        saveDC = mfcDC.CreateCompatibleDC()
        # 创建bigmap准备保存图片
        saveBitMap = win32ui.CreateBitmap()
        # 获取监控器信息
        MoniterDev = win32api.EnumDisplayMonitors(None, None)
        w = MoniterDev[0][2][2]
        h = MoniterDev[0][2][3]
    #     print(w,h)#图片大小
        if srcbmp[2]==None or (srcbmp[0]+srcbmp[2]>w):
            srcbmp[2] = w
        if srcbmp[3]==None or (srcbmp[1]+srcbmp[3]>w):
            srcbmp[3] = h
        # 为bitmap开辟空间
        saveBitMap.CreateCompatibleBitmap(mfcDC,srcbmp[2], srcbmp[3])
        # 高度saveDC，将截图保存到saveBitmap中
        saveDC.SelectObject(saveBitMap)
        # 截取从左上角（0，0）长宽为（w，h）的图片
        saveDC.BitBlt((0, 0), (w, h), mfcDC, (srcbmp[0], srcbmp[1]), win32con.SRCCOPY)
        saveBitMap.SaveBitmapFile(saveDC, filename)
    
    def Command4_Cmd(self, event=None):
        #TODO, Please finish the function here!
        top.destroy()

    def Command3_Cmd(self, event=None):
        #TODO, Please finish the function here!
        
        self.image=self.image_path+'/'+str(self.number)+'.jpg'
        self.window_capture(self.image,box)
        img_open = PIL.Image.open(self.image)
        img_resize = img_open.resize((400, 300),PIL.Image.ANTIALIAS)
        photo = PIL.ImageTk.PhotoImage(image=img_resize)
        self.Label1.config(image=photo)
        self.Label1.image = photo
        self.number +=1
#         number = 1
#         while(1):
#             window_capture(str(number)+".jpg",[443, 229, 1555, 873])
#             number += 1
#             time.sleep(5)
            

    def Command1_Cmd(self, event=None):
        #TODO, Please finish the function here!
        def on_mouse(event, x, y, flags, param):
            global img, point1, point2,box
            img2 = img.copy()
            if event == cv2.EVENT_LBUTTONDOWN:         #左键点击
                point1 = (x,y)
#                 cv2.circle(img2, point1, 10, (0,255,0), 5)
                cv2.imshow('image', img2)
            elif event == cv2.EVENT_MOUSEMOVE and (flags & cv2.EVENT_FLAG_LBUTTON):               #按住左键拖曳
                cv2.rectangle(img2, point1, (x,y), (0,0,255), 2)
                cv2.imshow('image', img2)
            elif event == cv2.EVENT_LBUTTONUP:         #左键释放
                point2 = (x,y)
                cv2.rectangle(img2, point1, point2, (0,0,255), 2) 
                cv2.imshow('image', img2)
                box = [point1[0],point1[1],point2[0]-point1[0],point2[1]-point1[1]]
                print(box)
                cv2.destroyAllWindows()
                
        self.window_capture('img/test.jpg')
        global img
        img = cv2.imread('img/test.jpg')
        cv2.namedWindow('image')
        cv2.setMouseCallback('image', on_mouse)
        cv2.imshow('image', img)
        cv2.waitKey(0)

    def Command2_Cmd(self, event=None):
        #TODO, Please finish the function here!
        self.number = 1
        self.image_path=tkFileDialog.askdirectory()  # 打开文件夹对话框
#         self.image_path = tkFileDialog.asksaveasfilename(title='选择图像保存路径', filetypes=[('JPG files', '.jpg','.png'), ('All Files', '*')])

if __name__ == "__main__":
    top = Tk()
    Application_Screen_image_acquisition(top).mainloop()

```
