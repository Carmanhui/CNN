```python
import os, sys
import numpy as np
import shutil
np.random.seed(2018)

if sys.version_info[0] == 2:
    from Tkinter import *
    from tkFont import Font
    from ttk import *
    #Usage:showinfo/warning/error,askquestion/okcancel/yesno/retrycancel
    from tkMessageBox import *
    #Usage:f=tkFileDialog.askopenfilename(initialdir='E:/Python')
    #import tkFileDialog
    #import tkSimpleDialog
else:  #Python 3.x
    from tkinter import *
    from tkinter.font import Font
    from tkinter.ttk import *
    from tkinter.messagebox import *
    import tkinter.filedialog as tkFileDialog
    #import tkinter.simpledialog as tkSimpleDialog    #askstring()

class Application_build_database(Frame):
    #这个类仅实现界面生成功能，具体事件处理代码在子类Application中。
    def __init__(self, master=None):
        Frame.__init__(self, master)
        self.master.title('制作数据集')
        self.master.geometry('492x381')
        self.createWidgets()
        self.train_samples = 0
        self.val_samples = 0

    def createWidgets(self):
        self.top = self.winfo_toplevel()

        self.style = Style()

        self.Text1Var = StringVar(value='Text1')
        self.Text1 = Entry(self.top, textvariable=self.Text1Var, font=('宋体',9))
        self.Text1.place(relx=0.309, rely=0.021, relwidth=0.652, relheight=0.087)

        self.style.configure('TCommand1.TButton', font=('宋体',12))
        self.Command1 = Button(self.top, text='选择原始数据', command=self.Command1_Cmd, style='TCommand1.TButton')
        self.Command1.place(relx=0.033, rely=0.021, relwidth=0.246, relheight=0.087)

        self.Text2Var = StringVar(value='Text2')
        self.Text2 = Entry(self.top, textvariable=self.Text2Var, font=('宋体',9))
        self.Text2.place(relx=0.309, rely=0.147, relwidth=0.652, relheight=0.087)

        self.style.configure('TCommand2.TButton', font=('宋体',12))
        self.Command2 = Button(self.top, text='选择保存路径', command=self.Command2_Cmd, style='TCommand2.TButton')
        self.Command2.place(relx=0.033, rely=0.147, relwidth=0.246, relheight=0.087)

        self.style.configure('TLabel1.TLabel', anchor='w', font=('宋体',12))
        self.Label1 = Label(self.top, text='测试数据比例', style='TLabel1.TLabel')
        self.Label1.place(relx=0.049, rely=0.315, relwidth=0.213, relheight=0.066)

        self.Text3Var = StringVar(value='0.8')
        self.Text3 = Entry(self.top, textvariable=self.Text3Var, font=('宋体',12))
        self.Text3.place(relx=0.293, rely=0.294, relwidth=0.181, relheight=0.087)

        self.style.configure('TCommand3.TButton', background='#3399FF', font=('宋体',12))
        self.Command3 = Button(self.top, text='开始制作', command=self.Command3_Cmd, style='TCommand3.TButton')
        self.Command3.place(relx=0.618, rely=0.294, relwidth=0.197, relheight=0.087)

        self.style.configure('TLabel2.TLabel', anchor='w', font=('宋体',12))
        self.Label2 = Label(self.top, text='训练样本类别', style='TLabel2.TLabel')
        self.Label2.place(relx=0.52, rely=0.483, relwidth=0.213, relheight=0.066)

        self.Text4Var = StringVar(value='5')
        self.Text4 = Entry(self.top, textvariable=self.Text4Var, font=('宋体',12))
        self.Text4.place(relx=0.764, rely=0.462, relwidth=0.181, relheight=0.087)

        self.style.configure('TLabel3.TLabel', anchor='w', font=('宋体',12))
        self.Label3 = Label(self.top, text='训练样本总数', style='TLabel3.TLabel')
        self.Label3.place(relx=0.52, rely=0.609, relwidth=0.213, relheight=0.066)

        self.Text5Var = StringVar(value='100')
        self.Text5 = Entry(self.top, textvariable=self.Text5Var, font=('宋体',12))
        self.Text5.place(relx=0.764, rely=0.588, relwidth=0.181, relheight=0.087)

        self.style.configure('TLabel4.TLabel', anchor='w', font=('宋体',12))
        self.Label4 = Label(self.top, text='测试样本类别', style='TLabel4.TLabel')
        self.Label4.place(relx=0.52, rely=0.735, relwidth=0.213, relheight=0.066)

        self.Text6Var = StringVar(value='5')
        self.Text6 = Entry(self.top, textvariable=self.Text6Var, font=('宋体',12))
        self.Text6.place(relx=0.764, rely=0.714, relwidth=0.181, relheight=0.087)

        self.style.configure('TLabel5.TLabel', anchor='w', font=('宋体',12))
        self.Label5 = Label(self.top, text='测试样本总数', style='TLabel5.TLabel')
        self.Label5.place(relx=0.52, rely=0.861, relwidth=0.213, relheight=0.066)

        self.Text7Var = StringVar(value='40')
        self.Text7 = Entry(self.top, textvariable=self.Text7Var, font=('宋体',12))
        self.Text7.place(relx=0.764, rely=0.84, relwidth=0.181, relheight=0.087)

        self.style.configure('TFrame1.TLabelframe', font=('宋体',12))
        self.style.configure('TFrame1.TLabelframe.Label', font=('宋体',12))
        self.Frame1 = LabelFrame(self.top, text='样本名称', style='TFrame1.TLabelframe')
        self.Frame1.place(relx=0.033, rely=0.42, relwidth=0.457, relheight=0.549)

        self.List1Var = StringVar(value='List1')
        self.List1Font = Font(font=('宋体',9))
        self.List1 = Listbox(self.Frame1, listvariable=self.List1Var, font=self.List1Font)
        self.List1.place(relx=0.071, rely=0.153, relwidth=0.858, relheight=0.766)

    def Command1_Cmd(self, event=None):
        #TODO, Please finish the function here!
        self.originaldata_dirname = tkFileDialog.askdirectory()  # 打开文件夹对话框
        self.Text1Var.set(self.originaldata_dirname)
        self.filename = os.listdir(self.originaldata_dirname)

    def Command2_Cmd(self, event=None):
        #TODO, Please finish the function here!
        self.savedata_dirname = tkFileDialog.askdirectory()  # 打开文件夹对话框
        self.Text2Var.set(self.savedata_dirname)

    def Command3_Cmd(self, event=None):
        #TODO, Please finish the function here!
        split_proportion = float(self.Text3Var.get())
        train_path = self.savedata_dirname + '/train'
        val_path = self.savedata_dirname + '/validation'
        os.mkdir(train_path)
        os.mkdir(val_path)
        for dc in self.filename:
            if dc not in os.listdir(train_path):
                os.mkdir(os.path.join(train_path, dc))
                
            total_images = os.listdir(os.path.join(self.originaldata_dirname, dc))
            split_number = int(len(total_images) * split_proportion)
            np.random.shuffle(total_images)
            train_images = total_images[:split_number]
            val_images = total_images[split_number:]
            for img in total_images:
                source = os.path.join(self.originaldata_dirname, dc, img)
                target = os.path.join(train_path, dc, img)
                shutil.copy(source, target)
                self.train_samples += 1

            if dc not in os.listdir(val_path):
                os.mkdir(os.path.join(val_path, dc))
                
            total_images = os.listdir(os.path.join(self.originaldata_dirname, dc))
            split_number = int(len(total_images) * split_proportion)
            np.random.shuffle(total_images)
            train_images = total_images[:split_number]
            val_images = total_images[split_number:]
            for img in val_images:
                source = os.path.join(self.originaldata_dirname, dc, img)
                target = os.path.join(val_path, dc, img)
                shutil.copy(source, target)
                self.val_samples += 1  
        
        #种类数       
        classes_number = len(self.filename)
        self.Text4Var.set(classes_number)
        self.Text6Var.set(classes_number)
        #训练数据每类的数量
        self.Text5Var.set(self.train_samples)
        #测试数据每类的数量
        self.Text7Var.set(self.val_samples)
        database_type = []
        database_type.append('train')
        for dc in self.filename:
            database_type.append('  '+dc)
        database_type.append('validation')
        for dc in self.filename:
            database_type.append('  '+dc)
        self.List1Var.set(database_type)
        
if __name__ == "__main__":
    build_database_top = Tk()
    Application_build_database(build_database_top).mainloop()
```
